```python
# 필요한 라이브러리들을 임포트합니다.
import os
from matplotlib import pyplot as plt
from PIL import Image
```

```python
# training 이미지와 annotation 미미지가 지정된 경로를 지정합니다.
PATH_TO_TEST_IMAGES_DIR = 'BRATS/images/training'
PATH_TO_ANNOTATION_DIR = 'BRATS/annotations/training'
# inch로 표현한 출력 이미지의 크기를 지정합니다.
IMAGE_SIZE = (12, 8)


# 디렉토리 경로와 이미지 이름을 합쳐서 최종적으로 이미지 팡리을 읽어올 경로를 지정합니다.
training_image_path = os.path.join(PATH_TO_TEST_IMAGES_DIR, 'VSD.Brain.XX.O.MR_T1c.54544_84.jpg')
annotation_image_path = os.path.join(PATH_TO_ANNOTATION_DIR, 'VSD.Brain.XX.O.MR_T1c.54544_84.png')

# training 이미지를 시각화합니다.
training_image = Image.open(training_image_path)
plt.figure(figsize=IMAGE_SIZE)
plt.imshow(training_image)

# annotation 이미지를 시각화합니다.
annotation_image = Image.open(annotation_image_path)

# annotation 이미를를 불러오고 시각화를 위해 0~5 사이의 레이블을 0~255사이의 픽셀값로로 변환해서 시각화합니다.
pix = annotation_image.load()
annotation_image_size = annotation_image.size
print(annotation_image_size)

for i in range(1,annotation_image_size[0]):
    for j in range (1,annotation_image_size[1]):
        pix[j,i] = int(round(pix[j,i]*63.75))

plt.figure(figsize=IMAGE_SIZE)
plt.imshow(annotation_image)
```
